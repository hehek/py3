№11

```python
def f1(x):
    return 5 * (x - 2)**4 - 1 / (x**2 + 8)
```

```python
def df1(x):
    return 20 * (x - 2)**3 + 2*x* (x**2 + 8)**(-2) #производная
```

```python
min = so.minimize(f1, 0.0, method='BFGS')
print('min f1(x) =', min.x) #менее точное вычисление
```

```python
print('Decision verification: f1(x) =', f1(min.x))
```

```python
min = so.minimize(f1, 0.0, method='BFGS', jac=df1) # jac — якобиан, в 1-мерном случае — производная
print('Min f1(x) =', min.x) #более точное вычисление
```

```python
print('Decision verification: f1(x) =', f1(min.x))
```

```python
def f2(x):
    return 4 * (x[0] - 3*x[1])**2 + 7 * x[0] ** 4
```

```python
min = so.minimize(f2, [5.0, 2.0], method='BFGS')
print('Min f2(x) =', min.x)
print('Decision verification: f2(x) =', f2(min.x))
```

№12

```python
def g1(x):
    return x**5 - 2*x**4 + 3*x - 7
```

```python
def g2(x):
    return x**5 + 2*x**4 - 3*x - 7
```

```python
x = np.linspace(-5, 5, 50)
g1_y = g1(x)
g2_y = g2(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, g1_y, 'xr') #красные кресты
plt.plot(x, g2_y, 'b') #синяя линия
plt.xlabel('x')
plt.ylabel('g1(x), g2(x)')
plt.legend(['g1', 'g2'])
plt.title('Plot')
plt.show()
```

№13

```python
plt.subplot(1, 4, 1)
plt.plot(x, g1_y, 'or')
plt.xlabel('x')
plt.ylabel('g1(x)')
plt.legend(['g1'])
plt.title('Plot №1')

plt.subplot(1, 4, 4)
plt.plot(x, g2_y)
plt.xlabel('x')
plt.ylabel('g2(x)')
plt.legend(['g2'])
plt.title('Plot №2')

plt.show()
```
